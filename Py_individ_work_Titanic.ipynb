```python
data_age = dataset['Age'].dropna()
df1 = pd.DataFrame(data_age)
bin_edges = [0, 15, 30, 45, 60, 75]
plt.hist(data_age, bins=bin_edges, color='grey', edgecolor='white')
plt.title('Histogram of Age of Titanic passengers')
plt.xlabel('Age')

# Set the x-ticks to specific intervals
plt.xticks(range(0, 90, 15))

plt.ylabel('Number of passengers')
plt.show()
```

```python
data_class = dataset['Pclass'].dropna()
df2 = pd.DataFrame(data_class)
pclass_counts = data_class.value_counts()

# Create a mapping dictionary for labels
label_mapping = {1: 'First Class', 2: 'Second Class', 3: 'Third Class'}

# Replace the labels
labels = [label_mapping[i] for i in pclass_counts.index]

plt.pie(pclass_counts, labels=labels, autopct='%1.0f%%')
plt.title('Pie chart of class of Titanic passengers')
plt.show()
```
